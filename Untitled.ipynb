```python
import tkinter as tk
from tkinter import messagebox
import random

class SpaceAdventureGame:
    def __init__(self, root):
        self.root = root
        self.root.title("Космическое приключение")
        
        self.spaceships = ["Звездный Истребитель", "Космический Фрегат", "Гравитационный Крейсер"]
        self.resources = 0
        self.enemies = ["Враждебный Дроид", "Космический Пират", "Черная Дыра", "Кибернетический Монстр"]
        self.create_widgets()
        
    def create_widgets(self):
        self.label = tk.Label(self.root, text="Добро пожаловать в Космическое приключение!", font=("Helvetica", 16))
        self.label.pack(pady=10)
        
        self.spaceship_var = tk.StringVar()
        self.spaceship_var.set(self.spaceships[0])
        self.spaceship_menu = tk.OptionMenu(self.root, self.spaceship_var, *self.spaceships)
        self.spaceship_menu.pack(pady=10)
        
        self.start_btn = tk.Button(self.root, text="Начать приключение", command=self.start_adventure)
        self.start_btn.pack(pady=10)
        
        self.result_label = tk.Label(self.root, text="", font=("Helvetica", 12))
        self.result_label.pack(pady=10)
        
    def start_adventure(self):
        self.resources = 0
        self.result_label.config(text="")
        spaceship = self.spaceship_var.get()
        
        messagebox.showinfo("Путешествие начинается", f"Вы отправились в путешествие на своем корабле {spaceship}.")
        
        for _ in range(5):
            event = random.choice(["resource", "enemy"])
            if event == "resource":
                resource_found = random.randint(5, 20)
                self.resources += resource_found
                messagebox.showinfo("Ресурсы обнаружены", f"Вы обнаружили астероид с ресурсами. Получено {resource_found} ресурсов.")
            elif event == "enemy":
                enemy = random.choice(self.enemies)
                messagebox.showinfo("Столкновение с врагом", f"Вы встретили {enemy}. Сражение начинается!")
                victory = random.choice([True, False])
                if victory:
                    victory_resources = random.randint(10, 30)
                    self.resources += victory_resources
                    messagebox.showinfo("Победа!", f"Вы победили {enemy} и получили дополнительные ресурсы: {victory_resources}.")
                else:
                    lost_resources = random.randint(5, 15)
                    self.resources -= lost_resources
                    messagebox.showinfo("Поражение", f"{enemy} оказался сильнее, ваш корабль поврежден. Потеряно {lost_resources} ресурсов.")
            messagebox.showinfo("Текущие ресурсы", f"Ваши текущие ресурсы: {self.resources}")
            
            if self.resources < 0:
                messagebox.showinfo("Конец приключения", "Ваши ресурсы закончились. Вы проиграли.")
                break
            
            if not messagebox.askyesno("Продолжение путешествия", "Желаете продолжить путешествие?"):
                break
        
        self.show_results()

    def show_results(self):
        if self.resources >= 100:
            self.result_label.config(text="Поздравляем! Вы собрали достаточно ресурсов и одержали победу в Космическом приключении.")
        else:
            self.result_label.config(text="К сожалению, вам не удалось собрать достаточно ресурсов. Попробуйте еще раз.")

if __name__ == "__main__":
    root = tk.Tk()
    game = SpaceAdventureGame(root)
    root.mainloop()
```
